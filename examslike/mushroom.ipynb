```python
import sys
import os
import pandas as pd
import numpy as np
from sklearn import preprocessing

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')   
```

```python
mushroom = pd.read_csv('agaricus-lepiota.data', header=None) #non c'Ã¨ header nel file
mushroom
```

```python
ordinalEnc = preprocessing.OrdinalEncoder()
mushroom_enc = pd.DataFrame(ordinalEnc.fit_transform(mushroom)) #il risultato di fit transform Ã¨ un array numpy
mushroom_enc
```

```python
X = pd.DataFrame(mushroom_enc.drop(columns=0,axis=1))
X
```

```python
Y = pd.DataFrame(mushroom_enc[0])
Y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

accuracy = []
svc_clf = SVC(kernel='linear', C=1)
scores = cross_val_score(svc_clf, X, Y, cv=5)
accuracy.append("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

knn_clf = KNeighborsClassifier(n_neighbors=5)
scores = cross_val_score(knn_clf, X, Y, cv=5)
accuracy.append("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

lr_clf = LogisticRegression(max_iter=500)
scores = cross_val_score(lr_clf, X, Y, cv=5)
accuracy.append("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

print(accuracy)
```
